```python
from tqdm import tqdm
tqdm.pandas()

def check_date_between(row):
    vgn_bgn_dttm, vgn_end_dttm, STND_YMD = row['vgn_bgn_dttm'], row['vgn_end_dttm'], row['STND_YMD']
    
    if pd.isna(vgn_bgn_dttm) or pd.isna(vgn_end_dttm):
        return 0
        
    if vgn_bgn_dttm <= STND_YMD <= vgn_end_dttm:
        return 1
    else:
        return 0
```

```python
sal_hist_df['vgn_YN'] = sal_hist_df.progress_apply(check_date_between, axis=1)
```
