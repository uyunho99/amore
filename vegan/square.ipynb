# DF 만들기

columns: 시간(일), 제품, 매출, group, post

- group: 평생 비건 인증 한번이라도 받음(1)
- post: 비건 유효기간 안에 있으면 (1)

```python
df = pd.merge(prd_mstr[['PRD_CD']],
              sal_prd_hist[['PRD_CD', 'STND_YMD', 'NET_PRD_SAL_AMT','SAL_UTPR']],
              on='PRD_CD', how='left')
```

```python
df
```

```python
new_df = df.groupby(['STND_YMD', 'PRD_CD'])[['NET_PRD_SAL_AMT', 'SAL_UTPR']].sum().reset_index()
```

```python
new_df
```

```python
new_df = new_df[new_df['PRD_CD'] != 'Z']
```

```python
new_df
```

```python
vgn_prd["PRD_CD"] = vgn_prd["PRD_CD"].astype(float).astype(int).astype(str)
```

```python
new_df = pd.merge(new_df, vgn_prd[['PRD_CD', 'vgn_bgn_dttm', 'vgn_end_dttm']],
              on='PRD_CD', how='left')
```

```python
new_df['group'] = np.where(new_df['vgn_bgn_dttm'].apply(lambda x: pd.to_datetime(x, errors='coerce'))>= pd.Timestamp('2019-01-01'), 1,0)
```

```python
# 날짜 datetime 형식으로 변환
new_df['STND_YMD'] = pd.to_datetime(new_df['STND_YMD'], errors='coerce')
new_df['vgn_bgn_dttm'] = pd.to_datetime(new_df['vgn_bgn_dttm'], errors='coerce')
new_df['vgn_end_dttm'] = pd.to_datetime(new_df['vgn_end_dttm'], errors='coerce')
```

```python
# post 추가: 사이에 있는지 확인
new_df['post'] = np.where(
    (new_df['STND_YMD'] >= new_df['vgn_bgn_dttm']) & (new_df['STND_YMD'] <= new_df['vgn_end_dttm']),
            1,
            0
)
```

```python
from tqdm import tqdm

target_ids = []

temp = new_df[new_df["group"] == 1]
ids = list(new_df["PRD_CD"].unique())

for i in tqdm(ids):
    t = len(temp[temp["PRD_CD"] == i]["post"].value_counts())
    if t == 2:
        target_ids.append(i)
        
target_ids
```

```python
result = pd.DataFrame(columns = new_df.columns)
for i in target_ids:
    temp = new_df[new_df["PRD_CD"] == i]
    result = pd.concat([result, temp], axis = 0)

result = result.reset_index(drop = True)
result
```
