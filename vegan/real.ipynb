1. 비어있는 날짜 선형보간법으로 채우기

```python
panel_real_incsno["STND_YMD"] = pd.to_datetime(panel_real_incsno["STND_YMD"])
```

```python
panel_real_incsno = panel_real_incsno.drop_duplicates(subset=["PRD_CD", "STND_YMD"], keep=False)
```

```python
# PRD_CD별로 누락된 날짜 채우기
def fill_missing_dates(group):
    date_range = pd.date_range(start=group["STND_YMD"].min(), end=group["STND_YMD"].max())
    group = group.set_index("STND_YMD").reindex(date_range).reset_index()
    group = group.rename(columns={"index": "STND_YMD"})
    group['PRD_CD'] = group['PRD_CD'].fillna(method='ffill').fillna(method='bfill')
    return group
```

```python
panel_real_incsno['STND_YMD'] = pd.to_datetime(panel_real_incsno['STND_YMD']) # 날짜 형식으로 변환
panel_real_incsno = panel_real_incsno.groupby("PRD_CD").apply(fill_missing_dates).reset_index(drop=True)
```

```python
# 선형보간 다시 수행
panel_real_incsno['NET_PRD_SAL_AMT'] =(
    panel_real_incsno.groupby('PRD_CD')['NET_PRD_SAL_AMT']
    .apply(lambda x: x.interpolate(method='linear'))
    .reset_index(drop=True)
)

panel_real_incsno['SAL_UTPR'] =(
    panel_real_incsno.groupby('PRD_CD')['SAL_UTPR']
    .apply(lambda x: x.interpolate(method='linear'))
    .reset_index(drop=True)
)
```

```python
# 소수점 첫째자리 반올림 ver

# STND_YMD 별로 NET_PRD_SAL_AMT와 SAL_UTPR의 평균을 계산하여 새로운 칼럼 추가
panel_real_incsno['avg_net_prd_sal_amt'] = panel_real_incsno.groupby('STND_YMD')['NET_PRD_SAL_AMT'].transform('mean')
panel_real_incsno['avg_sal_utpr'] = panel_real_incsno.groupby('STND_YMD')['SAL_UTPR'].transform('mean')

# time_demend1 및 time_demend2 칼럼 계산
panel_real_incsno['time_demend1'] = panel_real_incsno['NET_PRD_SAL_AMT'] - panel_real_incsno['avg_net_prd_sal_amt']
panel_real_incsno['time_demend2'] = panel_real_incsno['SAL_UTPR'] - panel_real_incsno['avg_sal_utpr']

# 소수점 첫째자리에서 반올림
panel_real_incsno['time_demend1'] = panel_real_incsno['time_demend1'].round(5)
panel_real_incsno['time_demend2'] = panel_real_incsno['time_demend2'].round(5)

# 불필요한 중간 칼럼 제거 (선택 사항)
panel_real_incsno.drop(columns=['avg_net_prd_sal_amt', 'avg_sal_utpr'], inplace=True)
```

2. txp FE 넣기

```python
# panel_real_incsno 테이블이 있다. 새로운 칼럼인 FE1, FE2를 만들것이다.
# FE1의 식은 다음과 같다. (time_demend1 컬럼) - 해당 제품의 NET_PRD_SAL_AMT 평균
# FE2의 식은 다음과 같다. (time_demend2 컬럼) - 해당 제품의 SAL_UTPR 평균
# 해당 제품의 평균은 PRD_CD에서 같은 제품의 평균을 구하면 된다.
# PRD_CD은 110030095 이런식으로 구성이 되어져있다. 같은 제품 별로 묶어서 NET_PRD_SAL_AMT를 내라는 소리이다.
```

```python
# PRD_CD 별로 time_demend1과 time_demend2의 평균값을 계산하여 새로운 칼럼 추가
panel_real_incsno['avg_time_demend1_by_prd'] = panel_real_incsno.groupby('PRD_CD')['time_demend1'].transform('mean')
panel_real_incsno['avg_time_demend2_by_prd'] = panel_real_incsno.groupby('PRD_CD')['time_demend2'].transform('mean')

# FE 칼럼 계산
panel_real_incsno['FE1'] = panel_real_incsno['time_demend1'] - panel_real_incsno['avg_time_demend1_by_prd'].round(5)
panel_real_incsno['FE2'] = panel_real_incsno['time_demend2'] - panel_real_incsno['avg_time_demend2_by_prd'].round(5)

# 불필요한 중간 칼럼 제거 (선택 사항)
panel_real_incsno.drop(columns=['avg_time_demend1_by_prd', 'avg_time_demend2_by_prd'], inplace=True)
```

```python
# 임시
import statsmodels.api as sm

panel_real_incsno["post"] = [1 if i >= 0 else 0 for i in panel_real_incsno["shock_distance"]]

x = panel_real_incsno["post"]
y = panel_real_incsno["FE1"]
x = sm.add_constant(x)

model = sm.OLS(y, x)
result = model.fit()
result.summary()
```
